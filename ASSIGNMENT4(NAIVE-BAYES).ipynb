```python
import numpy as np
import pandas as pd
```

## Diabetes dataset

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
df.isna().sum()
```

```python
X = df.drop("outcome", axis=1)
y = df["outcome"]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
np.random.seed(42)
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2)
bayes = GaussianNB()
bayes.fit(Xtrain, ytrain)
bayes.score(Xtest, ytest)
```

## Breast-cancer Dataset

```python
cancer = pd.read_csv("breast-cancer.csv")
cancer.head()
```

```python
cancer.isna().sum()
```

```python
cancer.drop("id", axis=1, inplace=True)
```

```python
cancer.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X = cancer.drop("diagnosis", axis=1)
y = cancer["diagnosis"]
```

```python
enc = LabelEncoder()
y_enc = enc.fit_transform(y)
print(y[:20])
print(y_enc[:20])
```

```python
np.random.seed(42)
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y_enc, test_size=0.2)
```

```python
bayes_cancer = GaussianNB()
bayes_cancer.fit(Xtrain, ytrain)
bayes_cancer.score(Xtest, ytest)
```
